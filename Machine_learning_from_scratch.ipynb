#Linear Regression from Scratch
##Gradient Descent

```python
import numpy as np

class LinearRegressionScratch:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # initialize parameters
        self.weights = np.zeros(n_features)
        self.bias = 0

        # gradient descent
        for _ in range(self.n_iters):
            y_pred = np.dot(X, self.weights) + self.bias

            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / n_samples) * np.sum(y_pred - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias


if __name__ == "__main__":
    X = np.array([[1], [2], [3], [4], [5]], dtype=float)
    y = np.array([2, 4, 6, 8, 10], dtype=float)

    model = LinearRegressionScratch(learning_rate=0.01, n_iters=2000)
    model.fit(X, y)

    predictions = model.predict(X)

    print("Weights:", model.weights)
    print("Bias:", model.bias)
    print("Predictions:", predictions)
```

##Normal Equation

```python
import numpy as np

class LinearRegressionNormalEquation:
    def __init__(self):
        self.weights = None

    def fit(self, X, y):

        # add bias column of 1s
        X_b = np.c_[np.ones((X.shape[0], 1)), X]

        # normal equation
        self.weights = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

    def predict(self, X):

        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        return X_b @ self.weights


# Example usage
if __name__ == "__main__":
    X = np.array([[1], [2], [3], [4], [5]], dtype=float)
    y = np.array([2, 4, 6, 8, 10], dtype=float)

    model = LinearRegressionNormalEquation()
    model.fit(X, y)

    predictions = model.predict(X)

    print("Weights (bias first, then slope):", model.weights)
    print("Predictions:", predictions)
```

##Locally Weighted Regression

```python
import numpy as np

X = np.array([1,2,3,4,5])
y = np.array([1,3,2,5,7])

x_query = 3
tau = 1

weights = np.exp(-(X-x_query)**2/(2*tau**2))

W = np.diag(weights)

X_b = np.c_[np.ones(len(X)), X]

theta = np.linalg.inv(X_b.T @ W @ X_b) @ X_b.T @ W @ y

prediction = theta[0] + theta[1]*x_query

print(prediction)
```
